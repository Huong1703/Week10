### Introduction:

This exercise was created based on the tutorial and documentation from [Seaborn](https://stanford.edu/~mwaskom/software/seaborn/index.html)  
The dataset being used is tips from Seaborn.

### Step 1. Import the necessary libraries:

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/tips.csv).

### Step 3. Assign it to a variable called tips

```python
tips = pd.read_csv('https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/tips.csv', sep=',')
print(tips)
```

### Step 4. Delete the Unnamed 0 column

```python
tips = tips.reset_index(drop=True)
print("Delete the Unnamed 0 column:")
print(tips)
```

### Step 5. Plot the total_bill column histogram

```python
plt.hist(tips['total_bill'], bins=20, edgecolor='black')
plt.title('total_bill Column')
plt.xlabel('Total Bill Amount')
plt.ylabel('Frequency')
plt.show()
```

### Step 6. Create a scatter plot presenting the relationship between total_bill and tip

```python
plt.scatter(tips['total_bill'], tips['tip'], edgecolor='black')
plt.title('Scatter Plot of total_bill vs tip')
plt.xlabel('Total Bill Amount')
plt.ylabel('Tip Amount')
plt.show()
```

### Step 7.  Create one image with the relationship of total_bill, tip and size.
#### Hint: It is just one function.

```python
plt.scatter(tips['total_bill'], tips['tip'], c=tips['size'], edgecolor='black', alpha=0.7)
plt.title('Scatter Plot of total_bill, tip, and size')
plt.xlabel('Total Bill Amount')
plt.ylabel('Tip Amount')
plt.colorbar(label='Size of Party')
plt.show()
```

### Step 8. Present the relationship between days and total_bill value

```python
import seaborn as sns
plt.figure(figsize=(8, 6))
ax = sns.barplot(x='day', y='total_bill', hue='day', data=tips, palette='muted', dodge=False)
ax.legend().set_visible(False)
plt.title('Relationship between Days and Total Bill')
plt.xlabel('Day of the Week')
plt.ylabel('Total Bill Amount')
plt.show()
```

### Step 9. Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='tip', y='day', hue='sex', data=tips, palette='viridis', edgecolor='black', s=100)
plt.title('Scatter Plot of Tip vs Day, Differentiated by Sex')
plt.xlabel('Tip Amount')
plt.ylabel('Day of the Week')
plt.legend(title='Sex')
plt.show()
```

### Step 10.  Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='day', y='total_bill', hue='time', data=tips, palette='Set2')
plt.title('Box Plot of Total Bill per Day, Differentiated by Time')
plt.xlabel('Day of the Week')
plt.ylabel('Total Bill Amount')
plt.legend(title='Time')
plt.show()
```

### Step 11. Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python
plt.figure(figsize=(12, 6))
sns.histplot(data=tips, x='tip', hue='time', multiple='stack', bins=20, palette='Set1', edgecolor='black')
plt.title('Histogram of Tip Value for Dinner and Lunch')
plt.xlabel('Tip Amount')
plt.ylabel('Frequency')
plt.legend(title='Time', loc='upper right')
plt.show()
```

### Step 12. Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker
### They must be side by side.

```python
g = sns.FacetGrid(tips, col='sex', hue='smoker', height=6)
g.map(plt.scatter, 'total_bill', 'tip', edgecolor='w', s=100, alpha=0.7)
g.add_legend(title='Smoker')
g.set_axis_labels('Total Bill Amount', 'Tip Amount')
g.set_titles(col_template="{col_name} Customers")
plt.show()
```
