### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Step 2. Create the DataFrame that should look like the one below.

```python
data = {
    'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
    'last_name': ['Miller', 'Jacobson', 'All', 'Milner', 'Cooze'],
    'age': ['42', '52', '36', '24', '73'],
    'female': ['0', '1', '1', '0', '1'],
    'preTestScore': ['4', '24', '31', '2', '3'],
    'postTestScore': ['25', '94', '57', '62', '70'],
}

data = pd.DataFrame(data)
print(data)
```

### Step 3. Create a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age
#### Hint: Don't forget to place the labels

```python
data['age'] = pd.to_numeric(data['age'])
data['preTestScore'] = pd.to_numeric(data['preTestScore'])
data['postTestScore'] = pd.to_numeric(data['postTestScore'])

plt.scatter(data['preTestScore'], data['postTestScore'], s=data['age']*5, alpha=0.5)
plt.xlabel('preTestScore')
plt.ylabel('postTestScore')
plt.title('Scatterplot of preTestScore and postTestScore (Size: Age)')
plt.show()
```

### Step 4. Create a Scatterplot of preTestScore and postTestScore.
### This time the size should be 4.5 times the postTestScore and the color determined by sex

```python
plt.scatter(data['preTestScore'], data['postTestScore'], s=4.5 * data['postTestScore'], c=data['female'], alpha=0.7)
plt.xlabel('preTestScore')
plt.ylabel('postTestScore')
plt.title('Scatterplot of preTestScore and postTestScore (Size: 4.5 times postTestScore, Color: Female)')
plt.colorbar(label='Female (0: Male, 1: Female)')
plt.show()
```
